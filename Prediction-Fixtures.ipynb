```python
import requests
import time
import requests
import json
import os

# Obtener IDs de fixtures
url = 'http://localhost:3000/api/matches/?select=id_fixture'
response = requests.get(url)
x = response.json()

def extraccion(record):
    return str(record["id_fixture"])

if "data" in x:
    ids = list(map(extraccion, x["data"]))
    print('* * * * **  ORIGINAL', len(ids))
else:
    print("No 'data' key found in the response.")

# Obtener IDs de predicciones
url_exists = 'http://localhost:3000/api/predictions-fixtures'
response_exists = requests.get(url_exists)
x_exists = response_exists.json()

def extraccion_exist(record):
    parameters = record["parameters"]
    return str(parameters["fixture"])

if "data" in x_exists:
    ids_exist = list(map(extraccion_exist, x_exists["data"]))
    print('** * * * * * EXISTENTES', len(ids_exist))
else:
    print("No 'data' key found in the response.")

def elementos_faltantes(array1, array2):
    faltantes = []
    for id in array1:
        if id not in array2:
            faltantes.append(id)
    return faltantes  

ids_completar = elementos_faltantes(ids, ids_exist)
print("******** COMPLETAR", len(ids_completar))



```
