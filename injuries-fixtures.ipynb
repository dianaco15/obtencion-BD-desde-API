```python
import requests
import json

url = 'http://localhost:3000/api/matches/?select=id_fixture'
response = requests.get(url)
x = response.json()

def extraccion(record):
    return str(record["id_fixture"])

# Filtrar y extraer solo los registros que tienen 'response' no vaco
if "data" in x:
    ids = [extraccion(record) for record in x["data"] if record.get("response")]
else:
    print("No 'data' key found in the response.")

url_exists = 'http://localhost:3000/api/injuries-fixtures/?select=parameters.fixture'
response_exists = requests.get(url_exists)
x_exists = response_exists.json()

def extraccion_exist(record):
    parameters = record["parameters"]
    return str(parameters["fixture"])

# Filtrar y extraer solo los registros que tienen 'response' no vaco
if "data" in x_exists:
    ids_exist = [extraccion_exist(record) for record in x_exists["data"] if record.get("response")]
else:
    print("No 'data' key found in the response.")

def elementos_faltantes(array1, array2):
    faltantes = []
    for id in array1:
        if id not in array2:
            faltantes.append(id)
    return faltantes  

ids_completar = elementos_faltantes(ids, ids_exist)
print("******** COMPLETAR", ids_completar, len(ids_completar))
```
