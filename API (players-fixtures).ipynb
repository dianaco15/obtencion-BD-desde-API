```python
import os
# URL de la API donde se enviarán los datos
post_url = 'http://localhost:3000/api/player-fixturesV2/'

# Obtener la ruta del archivo actual
current_directory = os.getcwd()

# Iterar a través de los partidos desde 143579 hasta 1200844
for fixture in ids_completar:
    
    
    
    
    # Construir la URL con los parámetros adecuados
    url = f"{url_base}?fixture={fixture}"
    
    # Hacer la solicitud
    response = requests.get(url, headers=headers)
    
    # Verificar si la solicitud fue exitosa
    if response.status_code == 200:
        data = response.json()
        
        # Guardar la respuesta en un archivo JSON independiente
        file_name = f"fixture_{fixture}.json"
        file_path = os.path.join(current_directory, file_name)
        with open(file_path, 'w') as json_file:
            json.dump(data, json_file, indent=4)
        
        print(f"Datos guardados para fixture {fixture} en {file_path}")
        
        # Enviar los datos a la API especificada
        post_response = requests.post(post_url, json=data)
        
        # Verificar si el POST fue exitoso
        if post_response.status_code == 200:  # Usualmente 201 para creación exitosa
            print(f"Datos enviados exitosamente para fixture {fixture} a la API.")
        else:
            print(f"Error al enviar datos para fixture {fixture} a la API: {post_response.status_code} - {post_response.text}")
    else:
        print(f"Error al obtener datos para fixture {fixture}: {response.status_code} - {response.text}")
    
    time.sleep(1)  # Adding a delay to avoid hitting rate limits
```
